## Create new features to use in algorithm.

```python
def create_features(df):
    df["sum_feature"] = df.iloc[:,2:202].sum(axis=1)
    df["mean_feature"] = df.iloc[:,2:202].mean(axis=1)
    df["std_feature"] = df.iloc[:,2:202].std(axis=1)
    df["Q1_feature"] = df.iloc[:,2:202].quantile(q =.25, axis=1)
    df["median_feature"] = df.iloc[:,2:202].median(axis=1)
    df["Q3_feature"] = df.iloc[:,2:202].quantile(q=.75, axis=1)
    df["skew_feature"] = df.iloc[:,2:202].skew(axis=1)
    df["kurtosis_feature"] = df.iloc[:,2:202].kurtosis(axis=1)
    
    # Maybe add difference between quantiles
    return df
```

```python
train_data_new = create_features(train_data.copy())
```

```python
train_data_new.head()
```

## Analyse new features to visualize any value

```python
sns.set_style("darkgrid")

# define 0 target and 1 target values
train_data_target_0 = train_data_new.loc[train_data_new['target'] == 0]
train_data_target_1 = train_data_new.loc[train_data_new['target'] == 1]

fig, axes = plt.subplots(2, 4, figsize=(30,15))
fig.suptitle("Distribution of new features in the train set", fontsize=20)

for idx in range(0,8):
    
    if idx < 4:
        N=0; M=idx
    else:
        N=1; M=idx-4
    
    sns.distplot(
        train_data_target_0.iloc[:,202+idx], 
        color="red", 
        kde=True, 
        bins=80, 
        label='target = 0',
        ax = axes[N][M]
    )
    sns.distplot(
        train_data_target_1.iloc[:,202+idx], 
        color="blue", 
        kde=True, 
        bins=80,
        label='target = 1',
        ax = axes[N][M]
    )
plt.show()
```

### New Features look very similar for both 0 and 1 tragets.
