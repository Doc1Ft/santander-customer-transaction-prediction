```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = os.path.join("..","data","train.csv")
train_data = pd.read_csv(csv_path)
```

```python
# Why is the mean target so low?
target_df = pd.DataFrame(train_data.target.value_counts())
target_df["Percentage"] = 100*target_df["target"]/train_data.shape[0]
target_df
```

```python
train_data_var = train_data.var().sort_values(ascending=True).reset_index().rename(columns={0:"Variance"})
```

```python
train_data_var.head()
```

```python
sns.set(font_scale=3)
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(20,10))
var_line = sns.lineplot(data=train_data_var["Variance"], ax=ax, linewidth=10)
ax.set_title("Variance (sorted)", fontsize=30)
ax.set_xticklabels("")
```

```python
train_data_std = train_data.std().sort_values(ascending=True).reset_index().rename(columns={0:"STD"})
```

```python
sns.set(font_scale=3)
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(20,10))
var_line = sns.lineplot(data=train_data_std["STD"], ax=ax, linewidth=10, color="red")
ax.set_title("Standard Deviation (sorted)", fontsize=30)
ax.set_xticklabels("")
```

## Take a look at some data transforms
- StandardScaler (Has mean of 0 and std of 1)
- MinMaxScaler (Values range from 0 to 1)
- QuantileTransformer (Assign each value to a quantile according to sample size)
- PowerTransform (mean of 0 and unit variance. Adjusts data to account for logarithmic patterns)
- Normalizer (Adjusts data to unit norm to l1 or l2.) mapped to unit circle
